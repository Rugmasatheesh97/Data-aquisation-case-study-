```python

import pandas as pd
df=pd.read_csv('/content/Electric_Vehicle_Data.csv.zip')
df.head()
```

```python
#  What is the size of the dataset? (Number of rows and columns)
dataset_size = df.shape
dataset_size
```

```python
#  What are the data types of each column?
column_data_types = df.dtypes
column_data_types
```

```python
# Are there any missing values in the dataset? If so, which columns have missing values and how many?
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values
```

```python
# Verify if 'Category' column exists
if 'Category' in df.columns:
    # Find unique values in the 'Category' column
    unique_values = df['Category'].unique()

    # Count the number of unique values
    num_unique_values = len(unique_values)

    print(f"Unique values: {unique_values}")
    print(f"Number of unique values: {num_unique_values}")
else:
    print("The 'Category' column is not present in the DataFrame.")
```

```python
# Check the data type of the 'VIN (1-10)' column
print(df['VIN (1-10)'].dtype)

# If it's not numeric, and if it makes sense to treat
# part of the VIN as numeric, extract the numerical part.
# Here, we assume the first two characters might be numeric.
# Adjust as needed based on your VIN structure.
df['VIN_Numeric'] = df['VIN (1-10)'].str[:2]
# Attempt to convert the extracted portion to numeric, handling errors
df['VIN_Numeric'] = pd.to_numeric(df['VIN_Numeric'], errors='coerce')

# Now try calculating the mean on the new numeric column
mean_value = df['VIN_Numeric'].mean()
median_value = df['VIN_Numeric'].median()
std_dev_value = df['VIN_Numeric'].std()

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_dev_value}")
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['VIN_Numeric'].quantile(0.25)
Q3 = df['VIN_Numeric'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['VIN_Numeric'] < lower_bound) | (df['VIN_Numeric'] > upper_bound)]

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: \n{outliers}")
```

```python
# Select only numeric columns before calculating correlation
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix on the numeric subset
correlation_matrix = numeric_df.corr()

print(correlation_matrix)
```
